```python
#ingest the file and test that it works
df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv("/Volumes/workspace/default/raw/transactions_raw.csv")
)
df.show(5)
og_rec_ct=df.count()
print(og_rec_ct)
```

```python
from pyspark.sql.functions import current_timestamp, lit

bronze_df = (
    df
    .withColumn("ingestion_timestamp", current_timestamp())
    .withColumn("source_system", lit("generated_csv"))
)
```

```python
(
    bronze_df.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("bronze_transactions")
)
```

```python
spark.sql("SELECT COUNT(*) FROM bronze_transactions").show()

display(spark.sql("SELECT * FROM bronze_transactions LIMIT 10"))
```

```python
df = spark.table("bronze_transactions")

# duplicates
df.groupBy("transaction_id").count().filter("count > 1").show()

# null prices
df.filter("price IS NULL").show()

# bad data
df.filter("price < 0").show()
```
